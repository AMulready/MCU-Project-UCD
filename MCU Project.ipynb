The box office is extremely varied, below we'll have a look at the results of Phase 1 of the MCU

```python
mcu_dict = {'Movie Title':      ['Iron Man','The Incredible Hulk','Iron Man 2','Thor','Captain America: The First Avenger',"Marvel's: The Avengers"],
              'IMDB Rating':    [7.9, 6.6, 7.0, 7.0, 6.9, 8.0],
           'Total Gross $': [585366247, 264770996, 623933331, 449326618, 370569774, 1518812988]}
number_with_commas = "{:,}".format(1000000)
mcu_dict
```

```python
mcu_df = pd.DataFrame(mcu_dict)
mcu_df
```

```python
plt.subplots(figsize=(10,10))
plt.xticks(rotation=45, ha='right')

plt.plot(mcu_df['Movie Title'], mcu_df['Total Gross $'])

plt.xlabel('Movie Titles')
plt.ylabel('Gross in $ (Billions)')
plt.title('Phase 1 Total Gross')
```

We can see here quite a difference in gross earnings. Iron Man was a new character to hit the big screen with lead actor Robert Downey Jr set to don the iconic suit, it's a financial success and so began the start of the Marvel Cinematic universe. The Incredible Hulk hits the screen, but due to this being a remake of the same character, becomes a box office flop. Iron Man 2 hits the screens in 2010 and performs slightly better than the original which is unusual for a sequel. Thor see his first big screen debut a year later to some success for a new character, but doesn't perform as well as Iron Man did 2 years earlier. Captain America is released a few months later, the lead actor, Chris Evans, having previously played a superhero from Marvel. Fans aren't as eager, but the film is still a success versus it's budget. The films so far have all lead up to a pivotal moment, the release of Marvel's: The Avengers, a box office smash generating over $1.5 billion worldwide.
